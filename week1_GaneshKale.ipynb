```python
import numpy as np
import pandas as pd


from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

```python
# array of size (10,50) with 500 random integers 
np.random.seed(101)
arr = np.array(np.random.randint(1,501,500).reshape(10,50))

# loadd array arr to dataframe using pandas

df = pd.DataFrame(arr)
df.head()
```

2. Calculate the sum of each row; the sum of each column; and the sum of all entries.

```python
# sum of each row:

row_sum = df.sum(axis=1)
row_sum

# sum of each columns:

col_sum = df.sum(axis=0)
col_sum

# sum of all entries:

df_sum = df.values.sum()  # df.to_numpy().sum()
df_sum
```

3. Determine the minimum; maximum; average; and standard deviation of all entries.

```python
print("DfataFrame Minimum: ",df.values.min())

print("DfataFrame Maximum: ",df.values.max())

print("DfataFrame Average: ",df.values.mean())

print("DfataFrame Standard Deviation: ",df.values.std())
```

4. Sort the DataFrame: a) sort on column 2; b) sort on row 2.

```python
# sort dataframe on column 2:

df.sort_values(by=2)
```

```python
# sort dataframe on row 2:

df.sort_values(by=2, axis=1)
```
